```python
import pandas as pd
```

```python
train = pd.read_csv('03_santander_train.csv', index_col='ID')
```

```python
train.head()
```

```python
test = pd.read_csv('03_santander_test.csv', index_col='ID')
```

```python
test.head()
```

```python
y = list(train['TARGET'])
```

```python
train = train.drop('TARGET', axis=1)
```

```python
train.info()
```

```python
train.info(verbose=True, memory_usage=True, null_counts=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, roc_curve
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.metrics import precision_recall_curve
```

```python
dtc = DecisionTreeClassifier(max_depth=5)
dtc.fit(X_train, y_train)
```

```python
from ipywidgets import Image
from io import StringIO
import pydotplus
from sklearn.tree import export_graphviz
import os
os.environ["PATH"] += os.pathsep + 'c:/Users/Ne0/Anaconda3/pkgs/graphviz-2.38.0-4/Library/bin/graphviz/'

dot_data = StringIO()
export_graphviz(dtc, feature_names=train.columns, out_file=dot_data, filled=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(value=graph.create_png())
```

```python
dtc = DecisionTreeClassifier(max_depth=1)
dtc.fit(X_train, y_train)
dtc_predict_proba_1 = dtc.predict_proba(X_test)

dtc = DecisionTreeClassifier(max_depth=2)
dtc.fit(X_train, y_train)
dtc_predict_proba_2 = dtc.predict_proba(X_test)

dtc = DecisionTreeClassifier(max_depth=3)
dtc.fit(X_train, y_train)
dtc_predict_proba_3 = dtc.predict_proba(X_test)

dtc = DecisionTreeClassifier(max_depth=4)
dtc.fit(X_train, y_train)
dtc_predict_proba_4 = dtc.predict_proba(X_test)

dtc = DecisionTreeClassifier(max_depth=5)
dtc.fit(X_train, y_train)
dtc_predict_proba_5 = dtc.predict_proba(X_test)
```

```python
plt.figure(figsize=(5, 5))
plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve DTC')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
fpr, tpr, thresholds = roc_curve(y_test, dtc_predict_proba_1[:,1])
plt.plot(fpr, tpr)
fpr, tpr, thresholds = roc_curve(y_test, dtc_predict_proba_2[:,1])
plt.plot(fpr, tpr)
fpr, tpr, thresholds = roc_curve(y_test, dtc_predict_proba_3[:,1])
plt.plot(fpr, tpr)
fpr, tpr, thresholds = roc_curve(y_test, dtc_predict_proba_4[:,1])
plt.plot(fpr, tpr)
fpr, tpr, thresholds = roc_curve(y_test, dtc_predict_proba_5[:,1])
plt.plot(fpr, tpr)
print(roc_auc_score(y_test, dtc_predict_proba_1[:,1]))
print(roc_auc_score(y_test, dtc_predict_proba_2[:,1]))
print(roc_auc_score(y_test, dtc_predict_proba_3[:,1]))
print(roc_auc_score(y_test, dtc_predict_proba_4[:,1]))
print(roc_auc_score(y_test, dtc_predict_proba_5[:,1]))
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
knn_predict_proba_1 = knn.predict_proba(X_test)

knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train, y_train)
knn_predict_proba_2 = knn.predict_proba(X_test)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
knn_predict_proba_3 = knn.predict_proba(X_test)

knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, y_train)
knn_predict_proba_4 = knn.predict_proba(X_test)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn_predict_proba_5 = knn.predict_proba(X_test)
```

```python
plt.figure(figsize=(5, 5))
plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve KNN')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
fpr, tpr, thresholds = roc_curve(y_test, knn_predict_proba_1[:,1])
plt.plot(fpr, tpr)
fpr, tpr, thresholds = roc_curve(y_test, knn_predict_proba_2[:,1])
plt.plot(fpr, tpr)
fpr, tpr, thresholds = roc_curve(y_test, knn_predict_proba_3[:,1])
plt.plot(fpr, tpr)
fpr, tpr, thresholds = roc_curve(y_test, knn_predict_proba_4[:,1])
plt.plot(fpr, tpr)
fpr, tpr, thresholds = roc_curve(y_test, knn_predict_proba_5[:,1])
plt.plot(fpr, tpr)
print(roc_auc_score(y_test, knn_predict_proba_1[:,1]))
print(roc_auc_score(y_test, knn_predict_proba_2[:,1]))
print(roc_auc_score(y_test, knn_predict_proba_3[:,1]))
print(roc_auc_score(y_test, knn_predict_proba_4[:,1]))
print(roc_auc_score(y_test, knn_predict_proba_5[:,1]))
```

```python
lr = LogisticRegression(penalty='l1', C=1.0)
lr.fit(X_train, y_train)
lr_predict_proba_1 = lr.predict_proba(X_test)

lr = LogisticRegression(penalty='l2', C=1.0)
lr.fit(X_train, y_train)
lr_predict_proba_2 = lr.predict_proba(X_test)

lr = LogisticRegression(penalty='l1', C=2.0)
lr.fit(X_train, y_train)
lr_predict_proba_3 = lr.predict_proba(X_test)

lr = LogisticRegression(penalty='l2', C=2.0)
lr.fit(X_train, y_train)
lr_predict_proba_4 = lr.predict_proba(X_test)

lr = LogisticRegression(penalty='l1', C=3.0)
lr.fit(X_train, y_train)
lr_predict_proba_5 = lr.predict_proba(X_test)
```

```python
plt.figure(figsize=(5, 5))
plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve LR')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
fpr, tpr, thresholds = roc_curve(y_test, lr_predict_proba_1[:,1])
plt.plot(fpr, tpr)
fpr, tpr, thresholds = roc_curve(y_test, lr_predict_proba_2[:,1])
plt.plot(fpr, tpr)
fpr, tpr, thresholds = roc_curve(y_test, lr_predict_proba_3[:,1])
plt.plot(fpr, tpr)
fpr, tpr, thresholds = roc_curve(y_test, lr_predict_proba_4[:,1])
plt.plot(fpr, tpr)
fpr, tpr, thresholds = roc_curve(y_test, lr_predict_proba_5[:,1])
plt.plot(fpr, tpr)
print(roc_auc_score(y_test, lr_predict_proba_1[:,1]))
print(roc_auc_score(y_test, lr_predict_proba_2[:,1]))
print(roc_auc_score(y_test, lr_predict_proba_3[:,1]))
print(roc_auc_score(y_test, lr_predict_proba_4[:,1]))
print(roc_auc_score(y_test, lr_predict_proba_5[:,1]))
```
